```python
import pandas as pd
from collections import Counter
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import math
from sklearn.metrics import r2_score
from tqdm import tqdm
```

```python
df = pd.read_pickle('Covid_meta_de.pkl')
df2 = pd.read_pickle('Covid_counts_de.pkl')
```

```python
uso=df2[df2.Count>3]
#uso1=df2[df2.Count>600]
#uso=uso1[uso1.Count<606]
uso.head()
#A coluna chama-se 'Tweet' nos dois ficheiros
```

```python
counts = uso['Count'].to_numpy()
print(counts)
```

```python
tweet_id = uso['Tweet'].to_numpy()
N=len(tweet_id)
print(tweet_id, N)
```

```python
uoso = df[df.Tweet==tweet_id[6]]
uoso.head()
#print(len(uoso))
```

```python
def FitMod(x, C, a, g, d):
    y = C/(np.exp((a/g)*np.exp(-1*(g*(x-d)))))
    return y
```

```python
for i in range(0, N, 1):
    try:
        f = open("de_cov_cascatas_teste4_0.01a.txt", "a")
    
        df_cas = df[df.Tweet==tweet_id[i]]
    
        time = df_cas['Time']
        size = np.arange(counts[i])+1
    
        xdata = np.asarray(time)
        ydata = np.asarray(size)
    
        w = len(xdata)//2
        shift = xdata[w]
    
        guess=[counts[i],0.01,0.0005,shift]
    
        parameters, covariance = curve_fit(FitMod, xdata, ydata, guess)
    
        fit_C = parameters[0]
        fit_a = parameters[1]
        fit_g = parameters[2]
        fit_shift = parameters[3]

        fit_y = FitMod(xdata, fit_C, fit_a, fit_g, fit_shift)
        r=r2_score(ydata, fit_y)
    
        L = [str(tweet_id[i]), " ",str(fit_C), " ", str(fit_a), " ", str(fit_g), " ", str(fit_shift), " ", str(r), "\n"]
        f.writelines(L)
    
        f.close()
    
    #plt.plot(xdata, ydata, 'o', label='data')
    #plt.plot(xdata, fit_y, '-', label='fit')
    #plt.legend()

    #ax[i].plot(xdata, ydata, label='data')
    #ax[i].plot(xdata, fit_y, label='fit')

#plt.show()
    except:
        ferr = open("de_cov_cascatas_err_teste4.txt", "a")
        sad = [str(tweet_id[i]), "\n"]
        ferr.writelines(sad)
        ferr.close()
        pass


```
