```python
import pandas as pd
import random
from random import choices
import pdb
from collections import Counter
import math

from scipy import stats
import arviz as az
import matplotlib.pyplot as plt
import numpy as np
import pymc3 as pm
from scipy.stats import gengamma
from datetime import datetime

#Observed data
data2 = pd.read_csv('cascade_distribution_de.csv', delim_whitespace=True)
y=data2.loc[:,"number"]
x=np.array(data2.loc[:,"Size"])
norm_y = y
max_val2 = max(y)
norm_y = norm_y/max_val2
max_x = max(data2.loc[:,"Size"])

print(len(norm_y))

def gamma_sim(form1, form2, scal1, scal2, scal3, scal4):
    itera = 3800
    gfg2=np.zeros(itera)
    cond = True
    
    while cond == True:
        for i in range(itera):
            gfg2[i] = round(gengamma.rvs(form1, form2, 0, scal1, size=1)[0])
        
        counter1 = Counter(gfg2)
        sizes1 = np.array(list(counter1.keys()))
        counts1 = np.array(list(counter1.values()))
        
        max_val = max(counts1)
        gfg2_norm1 = counts1/max_val
        cond_check = len(gfg2_norm1)
        x_gfg = sizes1[0:len(norm_y)]
        gfg = gfg2_norm1[0:len(norm_y)]*np.power((1+np.exp(-scal2*x_gfg)),scal3)/scal4
        print(siz, ' ', len(gfg))
        if cond_check>siz:
            cond = False
    print(siz, ' ', len(gfg))
    
    return x_gfg, gfg


f1 = 1
f2 = 0.735
s1 = 20
s2 = 0.15
s3 = 7.3
s4 = 170

x_gfg, gfg = gamma_sim(f1,f2,s1,s2,s3,s4)

print(len(norm_y), ' ', len(gfg))

plt.scatter(x, norm_y)
plt.scatter(x_gfg, gfg)
plt.xscale("log")
plt.yscale("log")
#plt.xlim([0,200])
#plt.ylim([0.00001,1])
```

```python
import numpy as np

def func1(x, val_i, const):
    val = val_i + const * (x+1)
    return val

def func2(x, val_i, val_f, const):
    vm = (val_i + const * (x+1))
    val= np.power(vm, 2)/val_f
    return val

def func3(x, val_i, val_f, const):
    vm = (val_i + const * (x+1))
    val= np.power(vm, 3)/(val_f*val_f)
    return val

val_i = 0
val_f = 0.3
num_steps = 10

const = (val_f - val_i)/num_steps

x_test = np.arange(num_steps)

vals1 = func1(x_test, val_i, const)

vals2 = func2(x_test, val_i, val_f, const)

vals3 = func3(x_test, val_i, val_f, const)

print("intervalo uniforme: ", vals1)
print("dependncia quadrtica: ", vals2)
print("dependncia cbica: ", vals3)
```
